```python
#import library
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#load and read data
dt = pd.read_csv('sample_data/breast-cancer.csv')
dt.head()
```

```python
dt.info()
```

```python
# Find the Null values in the dataframe and sum them
dt.isnull().sum()
```

```python
dt = dt.drop(["id"], axis = 1)
dt = dt.drop("Unnamed: 32", axis=1)
dt.head()
```

```python
G = dt[dt.diagnosis == "M"]
J = dt[dt.diagnosis == "B"]

plt.title("Tumor Ganas vs Tumor Jinak")
plt.xlabel("Rata-rata Radius")
plt.ylabel("Rata-rata Tekstur")
plt.scatter(G.radius_mean, G.texture_mean, color = "red", label = "Ganas", alpha = 0.3)
plt.scatter(J.radius_mean, J.texture_mean, color = "lime", label = "Jinak", alpha = 0.3)
plt.legend()
plt.show()
```

```python
dt.diagnosis = [i if i =="M" else 0 for i in dt.diagnosis]

x = dt.drop(["diagnosis"], axis = 1)
y = dt.diagnosis.values.astype(str)
```

```python
# Normalisasi
x = (x - np.min(x)/(np.max(x) - np.min(x)))

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
print("Naive Bayes score: ",nb.score(x_test, y_test))
```
