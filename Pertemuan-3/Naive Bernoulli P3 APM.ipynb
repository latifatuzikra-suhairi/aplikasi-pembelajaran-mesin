```python
# import library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# import dataset
dataset = pd.read_csv('NaiveBayes.csv')

# split data menjadi input dan output
X = dataset.iloc[:, [0,1]].values
y = dataset.iloc[:, 2].values
```

```python
# training dan testing data
from sklearn.model_selection import train_test_split

# menetapkan ukuran data : 25%
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size= 0.25, random_state=0)
```

```python
# import standard scaler
from sklearn.preprocessing import StandardScaler

# skala input data
sc_X = StandardScaler() 
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
# import classifier
from sklearn.naive_bayes import BernoulliNB

# inisialisasi NB
classifer = BernoulliNB()

# training model
classifer.fit(X_train, y_train)

# testing model
y_pred = classifer.predict(X_test)
```

```python
# import akurasi model
from sklearn.metrics import accuracy_score

# cetak akurasi
print(accuracy_score(y_pred, y_test))
```

```python
# visualisasi data
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
l = plt.axis()
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=20, cmap='RdBu', alpha=0.1)
plt.axis(l);
```
