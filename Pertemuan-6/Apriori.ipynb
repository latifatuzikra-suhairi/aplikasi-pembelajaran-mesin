```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# load and read data
data = pd.read_csv('gdrive/MyDrive/Aplikasi Pembelajaran Mesin/Pertemuan 6/Dataset/minimarket.csv',
                   delimiter=';', header=None)
data.head()
```

```python
data.info()
```

```python
data = data.values.astype(str)
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data) 
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
df = df.drop('nan', axis=1)
df.head()
```

```python
freq_itemsets = apriori(df, min_support=0.3, use_colnames=True)
freq_itemsets
```

```python
res = association_rules(freq_itemsets, metric='confidence', min_threshold=0.3)
res
```

```python
res1 = res[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
res1
```
