```python
#import library
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#load and read data
dt = pd.read_csv('sample_data/zoo.csv')
dt
```

```python
dt["animal"].unique()
```

```python
dt = dt.drop("uniq_id", axis=1)
dt.head()
```

```python
pd.get_dummies(dt.drop("animal", axis=1))
```

```python
pd.get_dummies(dt.drop("animal", axis=1), drop_first=True)
```

```python
X = pd.get_dummies(dt.drop("animal", axis=1), drop_first=True)
y = dt["animal"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=44)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
predictions
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10,8), dpi=150)
plot_tree(model, feature_names=X.columns, filled=True);
```
