```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#load and read data
dt = pd.read_csv('sample_data/salary.csv')
dt
```

```python
dt.info()
```

```python
#mencari dan menangani missing values
dt.isnull().sum()
```

```python
#membuat variabel x dan y.
x = dt['Experience_Years'].values.reshape(-1,1)
y = dt['Salary'].values.reshape(-1,1)

#split data menjadi training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

#object linear regresi
lin_reg = LinearRegression()
#training the model menggunakan training data yang sudah displit sebelumnya
model = lin_reg.fit(x_train, y_train)

#nilai slope/koefisien (m) dan intercept (b)
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
#accuracy score dari model kita menggunakan testing data yang sudah displit sebelumnya
lin_reg.score(x_test, y_test)
```

```python
#visualisasi Regression Line menggunakan data testing
plt.figure(figsize=(8,6))
plt.scatter(x_test, y_test)
plt.plot(x_test, model.predict(x_test), color='red')

plt.xlabel('Experience_Years')
plt.ylabel('Salary')
plt.title('Visualisasi Data (Testing)')
plt.show()
```

```python
#visualisasi menggunakan data training
plt.figure(figsize=(8,6))
plt.scatter(x_train, y_train)
plt.plot(x_train, model.predict(x_train), color = 'red')

plt.xlabel('Experience_Years')
plt.ylabel('Salary')
plt.title('Visualisasi Data (Training)')
plt.show()
```

```python
#memprediksi salary ketika experience year 12 tahun
Experience = np.array([20]).reshape(-1,1)

prediksi_salary = model.predict(Experience)
prediksi_salary
```
